# Singly Linked List

```python
class Node:
  def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None


class LinkedList:
  def __init__(self):
    self.headval = None

  # 1) Traversal / print
  def listprint(self):
    temp = self.headval
    while temp:
      print(temp.dataval)
      temp = temp.nextval

    # 2) Tambah di awal
  def AddBegining(self, newdata):
    newnode = Node(newdata)
    newnode.nextval = self.headval
    self.headval = newnode

    # 3) Tambah di tengah (setelah mid_node)
  def AddInbetween(self, mid_node, newdata):
    if mid_node is None:
      print("mid_node tidak boleh None")
      return
    newnode = Node(newdata)
    newnode.nextval = mid_node.nextval
    mid_node.nextval = newnode

  # 4) Tambah di akhir
  def AddInEnd(self, newdata):
    newnode = Node(newdata)
    if self.headval is None:
      self.headval = newnode
      return
    last = self.headval
    while last.nextval:
      last = last.nextval
    last.nextval = newnode

  # 5) Hapus berdasarkan nilai (key)
  def RemoveNode(self, key):
    temp = self.headval

    # Jika yang dihapus ada di head
    if temp and temp.dataval == key:
      self.headval = temp.nextval
      return

    prev = None
    while temp and temp.dataval != key:
      prev = temp
      temp = temp.nextval

    # Jika tidak ketemu
    if temp is None:
      return

    prev.nextval = temp.nextval

  # 6) Hapus node pertama
  def RemoveFirst(self):
    if self.headval:
      self.headval = self.headval.nextval

    # Bonus: Hapus node terakhir (tetap pakai logika "memutus rantai")
  def RemoveEnd(self):
    if self.headval is None:
      return
    if self.headval.nextval is None:
      self.headval = None
      return

    prev = self.headval
    temp = self.headval.nextval
    while temp.nextval:
      prev = temp
      temp = temp.nextval
    prev.nextval = None

# Inisialisasi linked list + 3 node bulan
ll = LinkedList()

n1 = Node("Januari")
n2 = Node("Februari")
n3 = Node("Maret")

# Susunan awal: Januari -> Februari -> Maret
ll.headval = n1
n1.nextval = n2
n2.nextval = n3

print("Linked List awal:")
ll.listprint()
```

```python
ll.AddBegining("Start")
print("Setelah AddBegining('Start'):")
ll.listprint()
```

```python
ll.AddInbetween(n2, "Middle")
print("Setelah AddInbetween(setelah 'Februari', 'Middle'):")
ll.listprint()
```

```python
ll.AddInEnd("The Last")
print("Setelah AddInEnd('The Last'):")
ll.listprint()
```

```python
ll.RemoveFirst()
print("setelah RemoveFirst():")
ll.listprint()
```

```python
ll.RemoveNode("Middle")
print("setelah RemoveNode('Middle'):")
ll.listprint()
```

```python
ll.RemoveEnd()
print("setelah RemoveEnd():")
ll.listprint()
```

# Latihan 2.1 Doubly Linked List
Silahkan pergi ke artikel saya dan scroll sampai ketemu Doubly Linked List

```python
class DNode:
  def __init__(self, data=None):
    self.data = data
    self.prev = None
    self.next = None

class DoublyLinkedList:
  def __init__(self):
    self.head = None
    self.tail = None

  def print_forward(self):
    cur = self.head
    while cur:
      print(cur.data, end=" <-> ")
      cur = cur.next
    print("None")

  def insert_after(self, after_node, data):
    if after_node is None:
      return

    newnode = DNode(data)
    newnode.prev = after_node
    newnode.next = after_node.next

    if after_node.next is None:
      self.tail = newnode
    else:
      after_node.next.prev = newnode

    after_node.next = newnode

  def remove_key(self, key):
    cur = self.head
    while cur and cur.data != key:
      cur = cur.next

    if cur is None:
      return

    if cur.prev is None:
      self.head = cur.next
    else:
      cur.prev.next = cur.next

    if cur.next is None:
      self.tail = cur.prev
    else:
      cur.next.prev = cur.prev

dll = DoublyLinkedList()

n1 = DNode("Januari")
n2 = DNode("Februari")
n3 = DNode("Maret")

dll.head = n1
dll.tail = n3

n1.next = n2
n2.prev = n1

n2.next = n3
n3.prev = n2
```

```python
# Sisip di tengah: setelah node "Februari"
mid = dll.head.next          # Januari -> (Februari)
dll.insert_after(mid, "SisipTengah")

print("Setelah sisip tengah (setelah Februari):")
dll.print_forward()

# Hapus dengan kata kunci
dll.remove_key("Februari")
print("Setelah hapus kata kunci 'Februari':")
dll.print_forward()
```

# Latihan 2.1 Circular Linked List
Silahkan pergi ke artikel saya dan scroll sampai ketemu Circular Linked List

```python
class CNode:
  def __init__(self, data=None):
    self.data = data
    self.next = None

class CircularLinkedList:
  def __init__(self):
    self.head = None

  def print_once_round(self):
    if self.head is None:
      print("Kosong")
      return

    cur = self.head
    while True:
      print(cur.data, end=" -> ")
      cur = cur.next
      if cur == self.head:
        break
    print("(kembali ke head)")

  def append(self, data):
    newnode = CNode(data)

    if self.head is None:
      self.head = newnode
      newnode.next = newnode
      return

    tail = self.head
    while tail.next != self.head:
      tail = tail.next

    tail.next = newnode
    newnode.next = self.head

  def remove_end(self):
    if self.head is None:
      return

    if self.head.next == self.head:
      self.head = None
      return

    prev = None
    cur = self.head

    while cur.next != self.head:
      prev = cur
      cur = cur.next

    prev.next = self.head

cll = CircularLinkedList()
cll.append("Januari")
cll.append("Februari")
cll.append("Maret")

print("Circular LL awal:")
cll.print_once_round()
```
